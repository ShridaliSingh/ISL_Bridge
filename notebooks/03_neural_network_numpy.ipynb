Neural Networks is an ML algorithm where we train a model based on some data, by adjusting weights and biases to give better predictions. When we say the model is becoming better, we mean that the value of the cost function is going down after each round of gradient descent. Once the training phase is over, the data is discarded and the weights and biases are fixed. As new data flows in, the network makes predictions based on these fixed parameters. This phase is refered to as inference.

```python
import numpy as np
```

```python
w1 = np.random.randn(4, 3) # weights connecting input to hidden
b1 = np.random.randn(4,) # biases for hidden layer
w2 = np.random.randn(2, 4) # weights connecting hidden to output
b2 = np.random.randn(2,) # biases for output layer
print("w1\n" ,w1)
print("b1\n" ,b1)
print("w2\n" ,w2)
print("b2\n" ,b2)
```

```python
data_train = np.random.randn(10,3)
a = np.array([0,1]) 
data_train_ans = np.repeat(a,5)
np.random.shuffle(data_train_ans)
print("data train\n", data_train)
print("data train ans\n",data_train_ans) 
```

```python
# Class 0 — points around (1, 1, 1)
X_class0 = np.random.randn(50, 3) + np.array([1, 1, 1])

# Class 1 — points around (-1, -1, -1)
X_class1 = np.random.randn(50, 3) + np.array([-1, -1, -1])

# Stack them together
data_train = np.vstack([X_class0, X_class1])
data_train_ans = np.array([0]*50 + [1]*50)
```

```python
w1 = np.random.randn(4, 3) * 0.01
b1 = np.random.randn(4) * 0.01
w2 = np.random.randn(2, 4) * 0.01
b2 = np.random.randn(2) * 0.01
```

```python
def forward(data_train,w1,b1,w2,b2):
    v1 = data_train @ w1.T + b1 
    a1 = np.maximum(0,v1)
    v2 = a1 @ w2.T + b2
    n = np.exp(v2)
    a2 = n / np.sum(n, axis = 1, keepdims = True)
    return a2 , a1 , v2, v1
```

```python
def cross_entropy_loss(data_train_ans, a2):
    indi_loss = -1*np.log(a2[np.arange(len(a2)), data_train_ans]) #shape = (10,1)
    loss = np.mean(indi_loss)
    return loss
```

```python
cross_entropy_loss(data_train_ans,a2)
```

```python
def one_hot(data_train_ans,n_classes):
    mat = np.zeros((len(data_train_ans),n_classes))
    mat[np.arange(len(data_train_ans)), data_train_ans] = 1
    return mat
one_hot(data_train_ans,2)
```

```python
def backprop(data_train, data_train_ans , a2, a1, v2, v1, w2,n_classes):
    one_hot_y = one_hot(data_train_ans,n_classes)
    dv2 = a2 - one_hot_y
    dw2 = dv2.T @ a1 / len(data_train_ans)
    db2 = np.mean(dv2, axis=0)
    da1 = dv2 @ w2
    dv1 = da1 * (v1 > 0)
    dw1 = dv1.T @ data_train / len(data_train_ans)
    db1 = np.mean(dv1, axis=0)
    return dw1,db1,dw2,db2
backprop(data_train, data_train_ans , a2, a1, v2, v1, w2,2)
```

```python
def update_params(w1, b1, w2, b2, dw1, db1, dw2, db2, learning_rate):
    w1 = w1 - learning_rate * dw1
    b1 = b1 - learning_rate * db1
    w2 = w2 - learning_rate * dw2
    b2 = b2 - learning_rate * db2
    return w1,b1,w2,b2
```

```python
def training(data_train,data_train_ans,w1,b1,w2,b2,n_classes):
    for _ in range(5000):
        a2,a1,v2,v1 = forward(data_train,w1,b1,w2,b2)
        loss = cross_entropy_loss(data_train_ans, a2)
        dw1,db1,dw2,db2 = backprop(data_train, data_train_ans , a2, a1, v2, v1, w2,n_classes)
        w1,b1,w2,b2 = update_params(w1, b1, w2, b2, dw1, db1, dw2, db2, 0.1)
        if _ % 100 == 0:
            print("loss : ", loss)
    return w1, b1, w2, b2
    
w1, b1, w2, b2 = training(data_train, data_train_ans, w1, b1, w2, b2,2)
```

```python
def accuracy(array1, array2):
    a = array1 == array2
    accu = (np.sum(a) / len(array1)) * 100
    return accu 
```

```python
a2,a1,v2,v1 = forward(data_train,w1,b1,w2,b2)
predictions = np.argmax(a2, axis=1)
accuracy(predictions, data_train_ans)
```
