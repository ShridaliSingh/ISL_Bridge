KNN - Kth nearest Neighbour.
This ML algorithm makes predictions based on a given dataset. It calculates the distance of the unknown quantity from all the point in the dataset using the euclidean distance formula and then amongst the k smallest distances, it takes a vote and predicts the value with majority as the unknown quantity.
This algorithm can be applied to any number of dimensions.

```python
x_train = np.array([[1,2,3],[-1,3,2],[2,5,0],[3,-3,1],[4,2,1]])
y_train = np.array(['X','Y','X','Z','W'])
x_test = np.array([[6,3,7]])
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
```

```python
def eucledian_distance(a,b):
    return np.sqrt(np.sum(np.square(a-b)))

a = np.array([1,2,3])
b = np.array([2,4,5])
print(eucledian_distance(a,b))
```

```python
distance = []
for ds in x_train:
    d = eucledian_distance(x_test[0],ds)
    distance.append(d)
distance = np.array(distance)
distance
```

```python
indices = np.argsort(distance)
indices
```

```python
k = 3 
label = []
for _ in range(k):
    label.append(y_train[indices[_]])
label
```

```python
k = 3
a = y_train[indices[:k]]
a
```

```python
from collections import Counter
counter = Counter(a)
b = counter.most_common(1)
b[0][0]
```

```python
def knn_predict(x_train,y_train,x_test,k):
    ldistance = []
    for tds in x_test:
        distance = []
        for ds in x_train:
            d = eucledian_distance(tds,ds)
            distance.append(d)
        ldistance.append(distance)
        list_distance = np.array(ldistance)

    ind = np.argsort(list_distance, axis = 1)
    label = y_train[ind[:,:k]]

    prediction = []
    for _ in label:
        counter = Counter(_)
        prediction.append(counter.most_common(1)[0][0])

    return prediction

x_train = np.array([[1,2,3],[-1,3,2],[2,5,0],[3,-3,1],[4,2,1]])
y_train = np.array(['X','Y','X','Z','W'])
x_test = np.array([[6,3,7],[1,1,1],[0,5,1]])
knn_predict(x_train,y_train,x_test,k = 3)
```
